```python
# Mean width_size: 473.51704545454544
# Mean height_size: 380.76799242424244

from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.resnet_v2 import preprocess_input
from PIL import Image
import tensorflow as tf

x_train = []
train_label = []
height_size=224
width_size=224

for image_label in train_list:
    filename = next((f'{image_label[0]}.{tail}' for tail in ['jpg', 'png', 'gif'] \
                 if os.path.isfile(f'{image_label[0]}.{tail}')), None)
    image = Image.open(filename)

    image = image.resize((height_size,width_size))

#     arr = arr.astype('float32') / 255.0
#     arr = np.array(image).astype('float32') / 127.5 - 1.0
    arr = np.array(image).astype('float32')
    arr = np.array(image).reshape(height_size, width_size, 3)

    x_train.append(arr)
    train_label.append(image_label[1])
x_train = np.array(x_train)
# _, width, height, channel = x_train.shape


```

```python
train_label= np.array(train_label)
y = np.array(train_label)

y = y.reshape(-1, 1)
y.shape
```

```python
x_test = []
test_id = []

for img_id in test_list:
    if img_id[:2] != '._':
        filename = f'/kaggle/input/dpl-fall2023/testset_images/testset_images/{img_id}'

        image = Image.open(filename)

        image = image.resize((height_size, width_size))

#         arr = arr.astype('float32') / 255.0
#         arr = np.array(image).astype('float32') / 127.5 - 1.0
        arr = np.array(image).astype('float32')
        arr = np.array(image).reshape(height_size, width_size, 3)
#         mean = np.mean(arr) 
#         arr -= mean
        
        x_test.append(arr)
        test_id.append(img_id[:-4]) 
x_test = np.array(x_test)
# _, wid, hei, chan = x_test.shape
```

```python
import tensorflow as tf
from keras.utils import to_categorical
y = tf.keras.utils.to_categorical(y)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train_img, X_dev_img, y_train, y_dev = train_test_split(x_train, y, test_size=0.1, random_state=42)
print(len(X_train_img))
print(len(X_dev_img))


```

```python
import tensorflow as tf
import tensorflow_ranking as tfr
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D,Dropout,BatchNormalization
import tensorflow.keras.optimizers as optimizers
from tensorflow.keras.applications import ResNet101V2,ConvNeXtXLarge
from tensorflow.keras.regularizers import l2,l1_l2
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import ModelCheckpoint,ReduceLROnPlateau
from sklearn.model_selection import train_test_split

X_train_img, X_dev_img, y_train, y_dev = train_test_split(x_train, y, test_size=0.1, random_state=42)

input_shape = (height_size, width_size, 3)

base_model = ConvNeXtXLarge(weights='imagenet', include_top=False, input_shape=input_shape,include_preprocessing=True)

for layer in base_model.layers[:-10]:
    layer.trainable = False

model = Sequential()

model.add(base_model)
model.add(GlobalAveragePooling2D())
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Dense(1024, activation='elu',kernel_regularizer=l2(0.01)))
model.add(BatchNormalization())
model.add(Dropout(0.7))

model.add(Dense(512, activation='elu',kernel_regularizer=l2(0.01)))
model.add(BatchNormalization())
model.add(Dropout(0.7))

model.add(Dense(256, activation='elu',kernel_regularizer=l2(0.01)))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Dense(128, activation='elu',kernel_regularizer=l2(0.01)))
model.add(BatchNormalization())
model.add(Dropout(0.5))

model.add(Dense(2, activation='sigmoid'))

optimizer = Adam(learning_rate=0.0001)
model.compile(loss='binary_focal_crossentropy', optimizer=optimizer, metrics=['accuracy'])

datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    brightness_range=(0.2, 0.8),
    horizontal_flip=True,
    vertical_flip=False,
    fill_mode='nearest'
)
batch_size = 64
augmented_data_generator = datagen.flow(X_train_img, y_train, batch_size=batch_size)

checkpoint = ModelCheckpoint(
    'best_model_checkpoint.h5',
    monitor='val_accuracy',
    save_best_only=True,
    mode='max',
    verbose=1
)
history = model.fit(
    augmented_data_generator,
    validation_data=(X_dev_img, y_dev),
    epochs=350,
    batch_size=batch_size,
    callbacks=[checkpoint]
)
model.save("/kaggle/working/last_train_model.h5")
```
